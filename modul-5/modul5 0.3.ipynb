```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

results = {}

# Define the 0.2% offset
offset_strain_value = 0.002

# Determine the number of datasets for subplot configuration
num_datasets = len(df)
# Adjust subplot layout if necessary, e.g., 6 columns for better aspect ratio if many plots
cols = 3 # Number of columns for subplots
rows = (num_datasets + cols - 1) // cols # Calculate rows needed

plt.figure(figsize=(5 * cols, 4 * rows)) # Adjust figure size based on cols and rows
plot_index = 1

for key, dataframe in df.items():
    if dataframe.empty or len(dataframe) < 2:
        print(f"Skipping {key} due to insufficient data.")
        # Ensure plot_index increments even if skipped, or handle subplotting differently
        # if plot_index <= rows * cols: plot_index +=1 
        continue

    # Determine material_type for lookup in initial_points_range
    material_type_for_lookup = None
    key_lower = key.lower()
    
    # Order checks from more specific to more general to correctly identify material type
    if "4_kertas_vertikal" in key_lower:
        material_type_for_lookup = "4_kertas_vertikal"
    elif "mika" in key_lower:
        material_type_for_lookup = "mika"
    elif "stik" in key_lower:
        material_type_for_lookup = "stik"
    elif "kertas" in key_lower:
        material_type_for_lookup = "kertas"
        
    if material_type_for_lookup is None:
        print(f"Warning: Could not map key '{key}' to a known material type for initial_points_range. Skipping.")
        # if plot_index <= rows * cols: plot_index +=1
        continue

    if material_type_for_lookup not in initial_points_range:
        print(f"Error: Material type '{material_type_for_lookup}' (from key '{key}') not in initial_points_range. Skipping.")
        # if plot_index <= rows * cols: plot_index +=1
        continue
        
    elastic_range_indices = initial_points_range[material_type_for_lookup]

    # Get strain and stress columns
    strain_col_name = dataframe.columns[0]
    stress_col_name = dataframe.columns[1]

    strain_raw = dataframe[strain_col_name].values
    stress_raw = dataframe[stress_col_name].values

    # Remove NaNs that might mess up sorting or calculations
    nan_mask = ~(np.isnan(strain_raw) | np.isnan(stress_raw))
    strain = strain_raw[nan_mask]
    stress = stress_raw[nan_mask]

    if len(strain) < 2:
        print(f"Skipping {key} after NaN removal, insufficient data.")
        # if plot_index <= rows * cols: plot_index +=1
        continue

    # Ensure data is sorted by strain
    sort_indices = np.argsort(strain)
    strain = strain[sort_indices]
    stress = stress[sort_indices]
    
    E = np.nan
    intercept_E = np.nan
    r_value_E = np.nan

    # --- Young's Modulus (E) ---
    # Ensure range is valid and has enough points
    if not (elastic_range_indices[0] >= 0 and elastic_range_indices[1] <= len(strain) and elastic_range_indices[0] < elastic_range_indices[1]):
        print(f"Warning for {key}: Invalid elastic_range_indices {elastic_range_indices} for data length {len(strain)}. E not calculated.")
    else:
        strain_elastic = strain[elastic_range_indices[0]:elastic_range_indices[1]]
        stress_elastic = stress[elastic_range_indices[0]:elastic_range_indices[1]]

        if len(strain_elastic) < 2:
            print(f"Warning for {key}: Not enough points ({len(strain_elastic)}) in elastic range for {material_type_for_lookup}. E not calculated.")
        else:
            E, intercept_E, r_value_E, _, _ = stats.linregress(strain_elastic, stress_elastic)
            if np.isnan(E):
                 print(f"Warning: Young's Modulus calculation failed for {key}. E is NaN.")

    # --- Ultimate Tensile Strength (UTS) ---
    if len(stress) == 0: # Should be caught by earlier checks
        uts_stress = np.nan
        uts_strain = np.nan
    else:
        uts_stress = np.max(stress)
        uts_strain_index = np.argmax(stress)
        uts_strain = strain[uts_strain_index]

    # --- Yield Strength (Ïƒ_y) using 0.2% offset ---
    yield_stress = np.nan
    yield_strain = np.nan
    found_yield = False

    if not np.isnan(E) and E > 0:
        search_start_index = elastic_range_indices[1] if len(elastic_range_indices) > 1 and elastic_range_indices[1] < len(strain) else 0
        
        for i in range(search_start_index, len(strain)):
            current_strain_val = strain[i]
            current_stress_val = stress[i]
            stress_on_offset_line = E * (current_strain_val - offset_strain_value) + intercept_E

            if current_stress_val >= stress_on_offset_line:
                # Try to interpolate with the previous point if it exists and was below the offset line.
                # This allows interpolation even if i == search_start_index.
                if i > 0: 
                    prev_strain_val = strain[i-1]
                    prev_stress_val = stress[i-1]
                    prev_stress_on_offset_line = E * (prev_strain_val - offset_strain_value) + intercept_E
                    
                    # Check if the curve segment (i-1 to i) crosses the offset line
                    if prev_stress_val < prev_stress_on_offset_line: # Previous point was below
                        # Interpolate for intersection
                        f1 = prev_stress_val - prev_stress_on_offset_line
                        f2 = current_stress_val - stress_on_offset_line
                        if (f2 - f1) != 0: # Avoid division by zero
                             t = -f1 / (f2 - f1)
                             if 0 <= t <= 1: # Intersection is within the segment
                                 yield_strain = prev_strain_val + t * (current_strain_val - prev_strain_val)
                                 yield_stress = prev_stress_val + t * (current_stress_val - prev_stress_val)
                                 found_yield = True
                                 break # Exit loop once yield is found by interpolation
                
                # If yield was found by interpolation, the loop breaks.
                # If not (e.g., i=0, or previous point wasn't below offset line, or interpolation failed),
                # found_yield remains False, and the loop continues to search for a valid intersection.
                # The direct assignment of current_strain_val/current_stress_val as yield point is removed
                # to strictly adhere to an interpolated intersection.
        
        if not found_yield:
            print(f"Yield point not identified for {key} using 0.2% offset method (no clear intersection with the offset line found).")

    results[key] = {
        "Young's Modulus (E)": E,
        "Intercept_E": intercept_E,
        "R_squared_E": r_value_E**2 if not np.isnan(r_value_E) else np.nan,
        "Yield Strength (Ïƒ_y)": yield_stress,
        "Yield Strain (Îµ_y)": yield_strain,
        "UTS (Ïƒ_u)": uts_stress,
        "Strain at UTS (Îµ_u)": uts_strain,
    }

    # --- Plotting ---
    if plot_index <= rows * cols:
        plt.subplot(rows, cols, plot_index)
        plot_index += 1

        plt.plot(strain * 100, stress, 'b-', label='Stress-Strain Curve', alpha=0.7, markersize=2, marker='.')

        if not np.isnan(E) and 'strain_elastic' in locals() and len(strain_elastic) > 0:
             plt.plot(strain_elastic * 100, E * strain_elastic + intercept_E, 'g--', linewidth=2, label=f"Elastic Fit (RÂ²={results[key]['R_squared_E']:.2f})")

        # Plot 0.2% offset line
        if not np.isnan(E) and E > 0 and not np.isnan(intercept_E):
            # Define strain points for the offset line
            # Start strain is offset_strain_value
            # End strain can be the maximum strain in the current dataset, 
            # or yield_strain if it's further (and valid), or a sensible fallback.
            
            strain_for_offset_plot_start = offset_strain_value
            strain_for_offset_plot_end = -1 # Initialize

            if len(strain) > 0:
                max_strain_data = np.max(strain)
                # Primarily, extend to max_strain_data if it's beyond the start of the offset line
                if max_strain_data > strain_for_offset_plot_start:
                    strain_for_offset_plot_end = max_strain_data
                # If yield_strain is found and is even further, use that (less common)
                if not np.isnan(yield_strain) and yield_strain > strain_for_offset_plot_end:
                    strain_for_offset_plot_end = yield_strain
            elif not np.isnan(yield_strain) and yield_strain > strain_for_offset_plot_start: 
                # Fallback if no 'strain' data but 'yield_strain' is valid and usable
                 strain_for_offset_plot_end = yield_strain
            
            # If no suitable end point found yet, use a small default extension for visibility
            if strain_for_offset_plot_end <= strain_for_offset_plot_start:
                 strain_for_offset_plot_end = strain_for_offset_plot_start * 1.5 # Arbitrary small extension if data is very limited

            # Ensure the end point is actually beyond the start point for a line to be drawn
            if strain_for_offset_plot_end > strain_for_offset_plot_start:
                offset_line_strains_to_plot = np.array([strain_for_offset_plot_start, strain_for_offset_plot_end])
                offset_line_stresses_to_plot = E * (offset_line_strains_to_plot - offset_strain_value) + intercept_E
                
                plt.plot(offset_line_strains_to_plot * 100, offset_line_stresses_to_plot, 
                         'k:', linewidth=1.5, label='0.2% Offset Line')

        if not np.isnan(yield_stress) and not np.isnan(yield_strain):
            plt.plot(yield_strain * 100, yield_stress, 'ro', markersize=7, label=f'Yield ({yield_stress:.2e} Pa)')

        if not np.isnan(uts_stress) and not np.isnan(uts_strain):
            plt.plot(uts_strain * 100, uts_stress, 'm*', markersize=9, label=f'UTS ({uts_stress:.2e} Pa)')
        
        title_parts = [key]
        if not np.isnan(E): title_parts.append(f"E={E:.2e}")
        if not np.isnan(yield_stress): title_parts.append(f"Yield={yield_stress:.2e}")
        if not np.isnan(uts_stress): title_parts.append(f"UTS={uts_stress:.2e}")
        plt.title("\n".join(title_parts), fontsize=8)
        plt.xlabel('Strain (%)', fontsize=8)
        plt.ylabel('Stress (Pa)', fontsize=8)
        plt.legend(fontsize='x-small')
        plt.grid(True)
        plt.tick_params(axis='both', which='major', labelsize=7)
    else:
        print(f"Skipping plot for {key} due to subplot limit.")


plt.tight_layout()
plt.show()

# Print results
print("\nCalculated Mechanical Properties:")
for key_res, props in results.items():
    print(f"\n--- {key_res} ---")
    print(f"  Young's Modulus (E): {props['Young\'s Modulus (E)']:.2e} Pa (RÂ²: {props['R_squared_E']:.3f})")
    if not np.isnan(props['Yield Strength (Ïƒ_y)']):
        print(f"  Yield Strength (Ïƒ_y) @ 0.2% offset: {props['Yield Strength (Ïƒ_y)']:.2e} Pa")
        print(f"  Yield Strain (Îµ_y): {props['Yield Strain (Îµ_y)']*100:.3f} %")
    else:
        print("  Yield Strength (Ïƒ_y): Not found or not calculated")
    print(f"  Ultimate Tensile Strength (UTS): {props['UTS (Ïƒ_u)']:.2e} Pa")
    print(f"  Strain at UTS (Îµ_u): {props['Strain at UTS (Îµ_u)']*100:.3f} %")
```
